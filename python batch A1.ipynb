```python
# LCM => Least Common Multiplier
# 11, 2
# 4, 7

x = 7
y = 4

if(x > y):
  greatest = x
else:
    greatest = y
while(True):
  if(greatest % x == 0 and greatest % y == 0):
    print("LCM is", greatest)
    break
  else:
    greatest += 1
```

```python
# program to check whether given number is an armstrong number

num = 145
length = len(str(num))
total = 0

while(num > 0):
  rem = num % 10
  total = total + rem ** length
  num = num // 10

if (total == 145):
  print("Num is an armstorng number")
else:
  print("Num is not an armstrong number")
```

```python
# program to check whether given number is a palindrome

num = 121
x = num
rev = 0

while(x > 0):
  rem = x % 10
  rev = rev * 10 + rem
  x = x // 10

if rev == num:
  print("Num is a palindrome")
else:
  print("Num is not a palindrome")
```

```python
# program to get all armstrong numbres between 100 and 400

for i in range(100,400):
  temp = i
  length = len(str(i))
  total = 0

  while(temp > 0):
    rem = temp % 10
    total = total + rem ** length
    temp = temp // 10

  if total == i:
    print(i)
```
